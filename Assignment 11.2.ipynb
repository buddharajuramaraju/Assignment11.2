##### Import required modules

```python
import numpy as np
import scipy as sp
from sklearn.utils import shuffle
from sklearn.feature_extraction.image import grid_to_graph
from sklearn.cluster import KMeans
from scipy.misc import face
from matplotlib import pyplot as pl
% matplotlib inline
```

```python
face = face(gray=True)
```

```python
pl.imshow(face,cmap=pl.cm.gray) # original image
```

```python
face.shape
```

##### clustering image

```python
number_of_clusters = 5
```

```python
#K-Means Clustering
kmeans = KMeans(n_clusters = number_of_clusters, n_init=10, max_iter=200)
```

```python
X = np.reshape(face, (-1, 1))
kmeans.fit(X)
```

```python
label = np.reshape(kmeans.labels_, face.shape)
print("Image after kmeans clustering")
pl.imshow(label,cmap=pl.cm.gray)
```

```python
print("Actual Image")
pl.imshow(face,cmap=pl.cm.gray)
```

##### Quality of image is reduced after clustering
